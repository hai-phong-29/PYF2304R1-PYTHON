```python
class Info:
    def __init__(self, name, age, phone, address):
        self.name = name
        self.age = age
        self.phone = phone
        self.address = address

    def __str__(self):
        return f"Name: {self.name}, age: {self.age}, phone: {self.phone}, address: {self.address}"


from tkinter import *
from tkinter import messagebox
from tkinter import ttk
import tkinter as tk

window = Tk()
window.title("Personal Information")
window.minsize(height=500, width=500)
window["bg"] = "#ffe4c4"

info_list = []

# Treeview
trv = ttk.Treeview(window)
trv.grid(row=8, column=1, padx=30, pady=30)

trv["columns"] = ("Name", "Age", "Phone", "Address")
trv['show'] = 'headings'
trv.column("Name", width=80, anchor='c')
trv.column("Age", width=80, anchor='c')
trv.column("Phone", width=80, anchor='c')
trv.column("Address", width=80, anchor='c')

trv.heading("Name", text="Name", anchor="c")
trv.heading("Age", text="Age", anchor="c")
trv.heading("Phone", text="Phone", anchor="c")
trv.heading("Address", text="Address", anchor="c")

trv.insert("", 'end', values=info_list)


def add_info():
    name_value = name.get()
    age_value = age.get()
    phone_value = phone.get()
    address_value = address.get()

    if name_value == "" or age_value == "" or phone_value == "" or address_value == "":
        messagebox.showerror("info", "Please enter name and age, phone, address!")
    else:
        new_info = Info(name_value, int(age_value), int(phone_value), address_value)
        info_list.append(new_info)
        print([info.name for info in info_list])
        trv.insert(
            "",
            "end",
            values=(name_value, int(age_value), int(phone_value), address_value),
        )
        messagebox.showinfo("Info", f"Added {name_value} successfully")


def search_info():
    search_info = search_name.get()
    check = False

    for i in range(len(info_list)):
        if info_list[i].name == search_info:
            messagebox.showinfo("Search Info", info_list[i])
            check = True
            break

    if check == False:
        messagebox.showerror("Search Info", f"Cannot find '{search_info}'!")


def delete_info():
    delete_info = search_name.get()
    check = False

    for i in range(len(info_list)):
        if info_list[i].name == delete_info:
            del info_list[i]
            print([info.name for info in info_list])
            messagebox.showinfo("Delete Info", f"Delete {delete_info} successfully!")
            check = True
            break
            

    if check == False:
        messagebox.showerror("Delete Info", f"Cannot find '{delete_info}'!")
        
    for n in trv.get_children():
            trv.delete(n)

    for info in info_list:
        trv.insert(
            "",
            "end",
            values=(name_value, int(age_value), int(phone_value), address_value)
        )


def update_infor():
    update_phone = search_name.get()
    new_phone = update_info.get()
    check = False
    for i in range(len(info_list)):
        if info_list[i].name == update_phone:
            info_list[i].phone = new_phone
            print([phone.name for phone in info_list])
            messagebox.showinfo("Update Phone", info_list[i])
            check = True
            break

    if check == False:
        messagebox.showerror("Update Phone", f"canot find '{update_phone}'!")
        
    for n in trv.get_children():
            trv.delete(n)

            for info in info_list:
                trv.insert(
                    "",
                    "end",
                    values=(name_value, int(age_value), int(phone_value), address_value)
                )
            


def reset():
    name.delete(0, "end")
    age.delete(0, "end")
    phone.delete(0, "end")
    address.delete(0, "end")
    search_name.delete(0, "end")
    update_info.delete(0, "end")


lbl_name = Label(window, text="Name: ")
lbl_age = Label(window, text="Age: ")
lbl_phone = Label(window, text="Phone: ")
lbl_address = Label(window, text="Address: ")
lbl_search = Label(window, text="Search: ")
lbl_update = Label(window, text="Update Phone: ")

name = Entry(window)
age = Entry(window)
phone = Entry(window)
address = Entry(window)
search_name = Entry(window)
update_info = Entry(window)

btn_add = Button(window, text="Add", command=add_info)
btn_update = Button(window, text="Update Phone", command=update_infor)
btn_delete = Button(window, text="Delete", command=delete_info)
btn_reset = Button(window, text="Reset", command=reset)
btn_search = Button(window, text="Search", command=search_info)


lbl_name.grid(row=0, column=0)
name.grid(row=0, column=1)
lbl_age.grid(row=1, column=0)
age.grid(row=1, column=1)
lbl_phone.grid(row=2, column=0)
phone.grid(row=2, column=1)
lbl_address.grid(row=3, column=0)
address.grid(row=3, column=1)
lbl_search.grid(row=5, column=0)
lbl_update.grid(row=6, column=0)

btn_add.grid(row=4, column=1)
btn_update.grid(row=6, column=2)
update_info.grid(row=6, column=1)
btn_delete.grid(row=7, column=2)
btn_reset.grid(row=7, column=0)
btn_search.grid(row=5, column=2)
search_name.grid(row=5, column=1)


window.mainloop()
```
