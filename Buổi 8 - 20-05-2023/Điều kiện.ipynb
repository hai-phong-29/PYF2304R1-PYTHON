## Câu lệnh if

```python
# viết chuowbg trình nhập vào giá trị đơn hàng của một khách hàng.
# Nếu giá trị đơn hàng lớn hơn 2tr thì in ra bạn được giảm giá 10%
total_price = int(input("Nhập giá trị đơn hàng: "))

if total_price > 2000000:
    print("Bạn được giảm giá 10%")
```

## Câu lệnh if..else

```python
# Viết một chương trình nhập vào điểm trung bình của một học sinh
# Nếu điểm trung bình lớn hơn 8, in ra học sinh giỏi
# Ngược lại, in ra "Bạn chwua phải học sinh giỏi"
average_score = float(input("Nhập điểm trung bình: "))
if average_score > 8:
    print("Bạn là học sinh giỏi")
else:
    print("Bạn không là học sinh giỏi")
    
```

### Thực hành:
Viết chương trình nhập vào một số nguyên và cho biết số nguyên đó là số chẵn hay số lẻ.

```python
number = int(input("Nhập vào số bất kì: "))
if number % 2 == 0:
    print("Là số chắn")
else:
    print("Là số lẻ")
```

### Thực hành:
Viết chương trình nhập vào một chuỗi và cho biết trong chuỗi đó có chứa cụm từ "python" hay không.

```python
string = input("nhập vào chuỗi bất kì: ")
if string.count('python') > 0:
    print(f"'python' exists in '{string}'")
else:
    print(f"'python' does not exist in '{string}'")
        
```

## Câu lệnh if lồng nhau

```python
total_hour = int(input("Please input total working hour of employee: "))
lever = int(input("Please input lever of employee: "))

if total_hour > 800:
    # khen thuong
    if lever == 1: 
        print("You are awarded 1000")
    else:
        print("You are awarded 2000")
else:
    print("Sorry bạn không được thưởng")
```

## Câu lệnh if…elif…else

```python
weekday = int(input("Nhập vào số ngày: "))
if weekday == 2:
    print("Monday")
    
elif weekday == 3: print("Tuesday")
    
elif weekday == 4:
    print("Wednesday")
    
elif weekday == 5:
    print("Thursday")

elif weekday == 6:
    print("Friday")

elif weekday == 7:
    print("Saturday")
    
else:
    print("Sunday")
```

### Thực hành: 
Viết chương trình nhập vào số tháng trong năm và cho biết tháng đó có bao nhiêu ngày

```python
month = int(input("Nhập một tháng bất kì: "))
if month == 1: print("31")
elif month == 2: print("28")
elif month == 3: print("31")
```
