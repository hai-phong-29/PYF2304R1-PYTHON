## Cách nhập xuất dữ liệu

### Xuất dữ liệu ra màn hình
Để in các nội dung mà chúng ta mong muốn ra màn hình, ta sử dụng lệnh print()

```python
# Ví dụ: In ra nội dung "python is very easy"
print("python easy")
```

### Các cách custom với lệnh print

Để in một nội dung kèm theo giá trị của biến, ta có các cách sau:

```python
food_1 = "cookies"
food_2 = "sandwich"

# I love chocolate and i hate sandwich
```

```python
# In trực tiếp
print("I love chocolate and i hate sandwich")
```

```python
# In bằng cách ghép các nội dung với nhau
print("I love", food_1, "and I hate", food_2, sep="+")
print("I love " + food_1 + " and i hate " + food_2)
```

```python
# Sử dụng String Format
print(f"I love {food_1} and I hate {food_2}")
```

```python
# Sử dụng nhiều lệnh print
print("I love ", end="")
print(food_1, end="")
print(" and I hate ", end="")
print(food_2, end="")
```

### Nhập dữ liệu
Để nhập dữ liệu vào chương trình python, chúng ta sử dụng hàm input()

```python
name = input("Please input your name: ")
print(name)
print(type(name))
```

```python
# type casting
number = float(input("Please input a number: "))
print(number)
print(type(number))
```

#### Thực hành: Viết chương trình nhập vào chiều dài và chiều rộng của hình chữ nhật, sau đó in ra chu vi và diện tích của hình chữ nhật đó

```python
# Nhập các kích thước của hình chữ nhật
chieu_dai = int(input("nhap chieu dai hinh chu nhat: "))
chieu_rong = int(input("nhap chieu rong hinh chu nhat: "))
```

```python
# Tính toán chu vi, diện tích
chu_vi = (chieu_dai + chieu_rong) * 2
dien_tich = chieu_dai * chieu_rong
```

```python
# In kết quả ra màn hình bằng 3 cách
print("chu vi la", chu_vi, "dien_tich la", dien_tich)
```

### Các thao tác phổ biến với string

```python
# Kiểm tra độ dài string
course_name = "Python Fundemental"
print(len(course_name))
```

```python
# Truy xuất ký tự bằng cơ chế indexing và reverse indexing
working_calendar = "oxoooxx"

# indexing
print(working_calendar[5])

# reverse indexing
print(working_calendar[-4])
```

#### Thực hành: Viết chương trình nhập vào một chuỗi bất kỳ, sau đó in ra ký tự đầu tiên và ký tự cuối cùng của chuỗi đó.

```python
mat_khau = "abcdefg"

print(mat_khau[0])
print(mat_khau[-1])
```

```python
# Cắt string theo index
# Cu phap: <ten bien>[start:end:step]
# default: start = 0, end = len(), step = 1

content = 'Long,Python,3-3-4,string'
print(content[0:4:])
print(content[5:11])
print(content[0:4:2])
print(content[18:24])
```

```python
# Ghép chuỗi và lặp chuỗi
a = "hello"
b = "world"
print(a + b) # ghep chuoi
print(a * 3) # lap chuoi
```
