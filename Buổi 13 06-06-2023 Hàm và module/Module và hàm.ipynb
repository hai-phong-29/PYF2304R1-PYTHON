# Function

## Function(hàm) là gì?

Hàm (function) là một khối lệnh đặc biệt, nó được đặt tên, giúp mã chương trình dễ đọc hơn, và có thể gọi để sử dụng ở các nơi khác nhau trong chương trình. Hàm chính là khối lệnh có thể tái sử dụng (reuseable).

### Hàm

Cú pháp:

`def <tên hàm>(<tham số)>:
    //Nội dung hàm`

Gọi hàm:
`<tên hàm>(<giá trị đầu vào>)`

### Ví dụ: Hàm in

```python
# viết hàm để giới thiệu về khóa học
def print_course_info():
    print("**********************************")
    print("*course: Python Fundamental      *")
    print("*Organization: Codegym Online    *")
    print("**********************************")
```

```python
print_course_info()
print_course_info()
```

### Ví dụ: Hàm tính giá trị tuyệt đối

```python
# viết hàm để in ra giá trị tuyệt đối của biến được input từ người dùng
def print_abs():
    number = int(input("nhập số: "))
    if number >= 0:
        print(f"Absolute value is {number}")
    else:
        print(f"Absolute value is {-number}")
```

```python
print_abs()
```

## Đối số của hàm

```python
# Hàm tính thể tích hình lập phương
def calc_cube_volume(side):
    print(f"cube volume is {side ** 3}")
    
calc_cube_volume(5)
calc_cube_volume(10)
calc_cube_volume(15)
```

```python
# Hàm xếp loại học sinh
def medium_score(a, b, c):
    medium = ((a + b + c) / 3)
    if medium >= 8:
        print("Học sinh giỏi")
    elif 6.5 < medium < 8:
        print("Học sinh khá")
    else:
        print("Học sinh trung bình")
medium_score(5, 4, 5)
```

```python
medium_score(10, 9, 9)
```

### Thực hành: Viết hàm nhận vào một số nguyên và kiểm tra xem số đó có phải là số nguyên tố hay không.


```python
def check_number(number):
    
    i_prime = True
    
    for i in range(2, number):
        if number % i == 0:
            print(f"{number} không phải là số nguyên tố")
            i_prime = False
            break
    if i_prime:
        print(f"{number} là số nguyên tố")

check_number(47)
```

# return

Cú pháp:

`return <giá trị trả về>`

##### Lưu ý: Sau khi thực hiện lệnh return, chương trình sẽ thoát khỏi hàm ngay lập tức (những lệnh sau đó không được thực hiện)

Ví dụ:

Hãy viết 1 hàm trả về giá trị trung bình của 2 số bất kỳ

```python
def calc_average(num1, num2):
    return ((num1 + num2) / 2)
    
result = calc_average(4, 5)

print(result)
```

```python
# Hàm tính chiều dài của chuỗi
# Viết hàm nhận vào một chuỗi và trả về chiều dài của chuỗi
def a(b):
    return len(b)
result = a('phongdo')

print(result)
```

```python
# Hàm tính giai thừa

def luythua(number):
    factorial = 1
    for i in range(2, number + 1):
        factorial *= i
    return factorial
print(luythua(5))
```

```python
# Hàm kiểm tra năm nhuận
# Viết hàm nhận vào một số năm và trả về True nếu năm đó là năm nhuận, ngược lại trả về False
def check_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False
```
