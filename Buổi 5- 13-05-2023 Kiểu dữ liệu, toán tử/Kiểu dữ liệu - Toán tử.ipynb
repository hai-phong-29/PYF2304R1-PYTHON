# Kiểu dữ liệu

```python
print(type(10))  # dạng số nguyên
print(type(5.5)) # dạng số thập phân
print(type("Hello Python")) # dạng chữ
print(type(True)) # dạng đúng/ sai
print(type(False)) # dạng đúng/ sai
```

```python
a = "Hello"
b = "Python"
c = a + b
print(c)
print(c * 3)
```

```python
# True -> 1, False -> 0
a = True
b = False
c = a + b
print(c)
```
